```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
image_color = mpimg.imread('/content/Chinese_Platter.jpg')
```

```python
plt.imshow(image_color);
```

```python
image_color.shape
```

```python
image_gray = cv2.cvtColor(image_color,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray');
```

```python
image_gray.shape
```

```python
image_gray
```

Saving the Gray scale image

```python
cv2.imwrite("My_Chinese_Platter.jpg",image_gray)
```
